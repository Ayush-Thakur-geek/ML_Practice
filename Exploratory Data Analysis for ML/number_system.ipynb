```python
from sympy import symbols, plot, Sum
from sympy.plotting import plot3d

x, y = symbols('x y')
f = 2*x + 1
plot(f)     
```

```python
f2 = 2*(x**3) + 1
plot(f2)     
```

```python
f3 =  2*(x**2) + 3*y
plot3d(f3)
```

```python
summation = sum(2*i for i in range(1,6))
print(summation)
```

```python
X = [1, 4, 7, 8]
n = len(X)
summation = sum(10*X[i] for i in range(0, n))
print(summation)
```

```python
i, n = symbols('i n')

summation = Sum(2 * i, (i, 1, n))

up_to_5 = summation.subs(n, 5)
print(up_to_5.doit())
```
